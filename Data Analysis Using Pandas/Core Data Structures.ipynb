<h1>Pandas is built on two main data structures:</h1>
<ol>
    <li>Series → One-dimensional (like a single column in Excel)</li>
    <li>DataFrame → Two-dimensional (like a full spreadsheet or SQL table)</li>
</ol>

<h3>Series — 1D Labeled Array</h3>
A Series is like a list with <strong>labels (index).</strong>

```python
import pandas as pd
s=pd.Series([100,244,66,55])
print(s)
```

Notice the automatic index: 0, 1, 2, 3

You can also define custom index:

```python
s=pd.Series([122,44,56,76],index=["Ansh","Anjali","Anu","Anju"])
print(s)
```

<h3>DataFrame — 2D Labeled Table</h3>
A DataFrame is like a <strong>dictionary of Series</strong> — multiple columns with labels.

```python
data={
    "name":["Ansh","Anju","Babu"],
    "age":[19,20,2],
    "city":["Punjab","Kota","Jamshedpur"]
}
df=pd.DataFrame(data)
print(df)
```

Each column in a DataFrame is a Series.

```python
df.index
```

```python
df.columns
```

<h3>Most Pandas operations are built on these foundations:</h3>
<ul>
    <li>Selection</li>
    <li>Filtering</li>
    <li>Merging</li>
    <li>Aggregation</li>
</ul>

Understanding Series & DataFrames will make everything else easier.

<h1>Summary</h1>
<ul>
    <li>Series = 1D array with labels</li>
    <li>DataFrame = 2D table with rows + columns</li>
    <li>Both come with index and are the heart of Pandas</li>
</ul>
