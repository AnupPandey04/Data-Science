<h5>Ranking</h5>
The .rank() function in pandas by default gives the average rank to tied values, which can result in decimal numbers. For example, if two people share the top score, they both get a rank of 1.5. You can customize the ranking behavior using the method parameter. One useful option is method='dense', which assigns the same rank to ties but doesnâ€™t leave gaps in the ranking sequence. This is helpful when you want a clean, consecutive ranking system without skips.

```python
df2["Rank"]=df2["IMDb"].rank()    #Default: Average rank ; for eg: if two tie then rank will be 1.5 for both
df2
```

```python
df2["Rank"]=df2["IMDb"].rank(method="dense")    #using method dense will not give rank in decimals
df2
```

<h1>Melt() & Pivot()</h1>

<h3>1. melt() â€” Wide to Long</h3>
The melt() method in Pandas is used to unpivot a DataFrame from wide format to long format.<br>
Syntax:

Parameters:<ul>
    <li>id_vars: The columns that you want to keep fixed (these columns will remain as identifiers).</li>
    <li>value_vars: The columns you want to unpivot (the ones you want to "melt" into a single column).</li>
    <li>var_name: The name to use for the new column that will contain the names of the melted columns (default is 'variable').</li>
    <li>value_name: The name to use for the new column that will contain the values from the melted columns (default is 'value').</li>
    <li>col_level: Used for multi-level column DataFrames.</li>
</ul>

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Math': [85, 78, 92],
    'Science': [90, 82, 89],
    'English': [88, 85, 94]
}
df = pd.DataFrame(data)
df   #Example df
```

```python
df.melt(id_vars="Name",value_vars=["Math","Science","English"],var_name="Subject",value_name="Score")
```

Why Use melt()?
<ul>
    <li>Data normalization: Helps in transforming data for statistical modeling and data visualization.</li>
    <li>Pivot tables: Many times, plotting functions or statistical models work better with long-format data.</li>
</ul>

<h3>2. pivot() â€” Long to Wide</h3>

The pivot() function in Pandas is used to reshape data, specifically to turn long-format data into wide-format data. <br>
This is the reverse operation of melt().<br>
Syntax:

Parameters:
<ul>
    <li>index: The column whose unique values will become the rows of the new DataFrame.</li>
    <li>columns: The column whose unique values will become the columns of the new DataFrame.</li>
    <li>values: The column whose values will fill the new DataFrame. These will become the actual data (values in the table).</li>
</ul>

```python
data = {
    'Name': ['Alice', 'Alice', 'Alice',
             'Bob', 'Bob', 'Bob',
             'Charlie', 'Charlie', 'Charlie'],
    'Subject': ['Math', 'Science', 'English',
                'Math', 'Science', 'English',
                'Math', 'Science', 'English'],
    'Score': [85, 90, 88,
              78, 82, 85,
              92, 89, 94]
}
df2 = pd.DataFrame(data)
df2    #Example df
```

```python
df.pivot(index="Name", columns="Subject", values="Score")
```

Why use pivot()?
<ul>
    <li>Better data structure: It makes data easier to analyze when you have categories that you want to split into multiple columns.</li>
    <li>Easier visualization: Often, you want to represent data in a format where categories are split across columns (for example, when creating pivot tables for reporting).</li>
    <li>Aggregating data: You can perform aggregations (like sum, mean, etc.) to group values before pivoting.</li>
</ul>

<h3>Important Notes:</h3>
<strong>1. Duplicate Entries:</strong> If you have multiple rows with the same combination of index and columns, pivot() will raise an error. <br>
In such cases, you should use pivot_table() (which can handle duplicate entries by aggregating them).

```python
data = {
    'Name': ['Alice', 'Alice', 'Alice', 'Bob', 'Bob'],
    'Subject': ['Math', 'Math', 'Science', 'Math', 'Math'],
    'Score': [85, 80, 90, 78, 82]
}
df = pd.DataFrame(data)
df
```

We can use pivot_table() to aggregate values (e.g., taking the mean for duplicate entries):

```python
df.pivot_table(index="Name", columns="Subject", values="Score", aggfunc="mean")
```

In this case, the Math score for Alice is averaged (85 + 80) / 2 = 82.5. If a cell is empty, it means there was no value for that combination.


<h1>Summary:</h1>
<ul>
    <li>Use melt() to go long, pivot() to go wide</li>
    <li>pivot() is used to turn long-format data into wide-format by spreading unique column values into separate columns.</li>
    <li>If there are duplicate values for a given combination of index and columns, you should use pivot_table() with an aggregation function to handle the duplicates.</li>
</ul>
