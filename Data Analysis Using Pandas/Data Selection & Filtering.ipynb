```python
import pandas as pd
df=pd.read_csv("data selection and filtering.csv")
df
```

<h2>Selecting Columns</h2>

```python
df["Film"]   # Single column (as Series)
```

```python
df[["Film","Year"]]       # Multiple column (as DataFrame)
```

<h2>Selecting Rows by Index</h2>
Use .loc[] (label-based) and .iloc[] (index-based):

```python
df.loc[0]   #First row (by label)
```

```python
df.iloc[0]      #First row (by position)
```

<h2>Select Specific Rows and Columns</h2>

```python
df.loc[0,"Actor"]     # Value at row 0, column "Actor"
```

```python
df.iloc[0,0]          # Value at row 0, column at index 1
```

We can also slice:

```python
df.loc[0:2,["Actor","Film"]]      # Fast label-based access
```

```python
df.iloc[0:2,0:2]                # Fast index based access
```

<h2>Filtering with Conditions</h2>

<h4>Simple Condition</h4>

```python
df[df["Year"]>2018]
```

<h4>Multiple Conditions (AND / OR)</h4>

```python
df[(df["IMDb"]>6.5) & (df["Genre"]=="Action")]
```

```python
df[(df["Film"]=="War") | (df["Actor"]=="Hritik Roshan")]
```

<h2>Querying with .query()</h2>

The .query() method in pandas lets you filter DataFrame rows using a string expression — it's a more readable and often more concise alternative to using boolean indexing.

This is a cleaner, SQL-like way to filter:

```python
df.query("IMDb>6 or Film =='War'")
```

```python
col="Year"
df.query(f"{col}<=2018")
```

<h4>rules and tips for using .query() in pandas:</h4><br>
<h4>1. Column names become variables</h4><br>
You can reference column names directly in the query string:

```python
df.query("Actor=='Salman Khan'")
```

<h4>2. String values must be in quotes</h4><br>
Use single or double quotes around strings in the expression:

```python
df.query("Actor=='Salman Khan'")
```

<h4>3. Use backticks for column names with spaces or special characters</h4>
If a column name has spaces, use backticks (`):

```python
df.query("`BoxOffice(INR Crore)`>400")
```

<h4>4. You can use @ to reference Python variables</h4><br>
To pass external variables into .query():

```python
a=6.0
df.query("IMDb==@a")
```

<h4>5. Logical operators</h4><br>
Use these:
<ul>
    <li>and, or, not — instead of &, |, ~</li>
    <li>==, !=, <, >, <=, >=</li>
</ul>
Bad:
df.query("age > 30 & city == 'Delhi'")   ❌

Good:
df.query("age > 30 and city == 'Delhi'")   ✅

<h4>6. Chained comparisons</h4>

```python
df.query("5<IMDb<=7")
```

<h4>7. Avoid using reserved keywords as column names</h4><br>
If you have a column named class, lambda, etc., you’ll need to use backticks:<br>
df.query("`class` == 'Physics'")

<h4>8. Case-sensitive</h4><br>
Column names and string values are case-sensitive:<br>
df.query("City == 'delhi'")   ❌ if actual value is 'Delhi'

<h4>9. .query() returns a copy, not a view</h4><br>
The result is a new DataFrame. Changes won't affect the original unless reassigned:

```python
newdf=df.query("IMDb>5")
newdf
```

<h2>Summary</h2><br>
<ul>
    <li>Use df[col], .loc[], .iloc[], .at[], .iat[] to access data</li>
    <li>Filter with logical conditions or .query() for readable code</li>
</ul><br>
In pandas, .at and .iat are fast accessors used to quickly get or set a single value in a DataFrame or Series.
